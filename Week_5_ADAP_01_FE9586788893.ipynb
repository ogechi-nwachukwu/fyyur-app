##6

```python
#Identify numerical columns with missing values and fill them using the mean

df.isnull().sum()
```

```python
df['experience'] = df['experience'].replace({'>20': 21,
                                             '<1': 0,'nan':0,
                                             'inf':0}).astype(float)
```

```python
df['experience'] = df['experience'].fillna(df['experience'].mean())
```

```python
df['last_new_job'].unique()
```

```python
df['last_new_job'] = df['last_new_job'].replace({'>4': 3,'never':0,
                                             'nan':0}).astype(float)
```

```python
df['last_new_job'] = df['last_new_job'].fillna(df['last_new_job'].mean())
```

```python
df['company_size'].unique()
```

```python
df['company_size'] = df['company_size'].replace({
    '<10': 9,
    '100-500': 500,
    '500-999': 999,
    '1000-4999': 4999,
    '5000-9999': 9999,
    '10000+': 10000,
    '50-99': 99,
    'Oct-49': 49,
    ' NaN' : 0,
    'nan': 0
}).astype(float)
```

```python
df['company_size'] = df['company_size'].fillna(df['company_size'].mean())
```

```python
df.isnull().sum()
```

##7


```python
#Drop rows where more than 50% of data is missing, and document the impact on dataset size.
threshold = df.shape[1] * 0.5
df_cleaned = df[df.isnull().sum(axis=1) <= threshold]

new_rows = df_cleaned.shape[0] - df.shape[0]
print(f"Number of rows dropped: {new_rows}")
```

There was no row with up to 50% missing data, so, no row was dropped

##10

```python
#Normalize City_Development_Index to a 0–1 scale and explain the benefits of normalization.
scaler = MinMaxScaler()
df['city_development_index'] = scaler.fit_transform(df[['city_development_index']])
df.head()
```

##11

```python
#Create new column cdi_per and merge it to the original dataframe using pd.merge and analyze the combined insights.

cdi_per = pd.DataFrame({
    'enrollee_id': df['enrollee_id'],  # Include enrollee_id
    'cdi_per': df['city_development_index'] * 100
})

merged_df = pd.merge(df, cdi_per, how='left')

print(merged_df.columns)
```
